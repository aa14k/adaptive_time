```python
import numpy as np
```

```python
def integrate(xs, tol, idxes):
    N = len(xs[0])
    if N > 2:
        Q = N * (xs[0,0] + xs[0,-1]) / 2
        idxes[int(xs[1,0])] += 1
        idxes[int(xs[1,-1])] += 1
    else:
        idxes[int(xs[1,0])] += 1
        idxes[int(xs[1,-1])] += 1
        return sum(xs[0])
    truth = sum(xs[0])
    if np.abs(Q - truth) > tol:
        c = int(np.floor(N / 2))
        Q = integrate(xs[:,:c], tol / 2, idxes) + integrate(xs[:,c:], tol / 2, idxes)
    return Q

    
```
